```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos desde el archivo CSV
datos = pd.read_csv('udd.csv')

# Ajuste de columnas
x = datos['¿Qué tanto influye el dinero y/o situación económica en tu felicidad?']
y = datos['Felicidad ¿Qué tan feliz eres actualmente?']

# Crear el gráfico de densidad (KDE Plot)
plt.figure(figsize=(10, 6))
sns.kdeplot(x=x, y=y, cmap='Blues', fill=True, thresh=0.05)

# Personalizar el gráfico
plt.title('Densidad de la Influencia de la Situación Económica en la Felicidad')
plt.xlabel('Influencia Económica')
plt.ylabel('Nivel de Felicidad')
plt.grid(True)

# Limitar el rango de los ejes a 0-10
plt.xlim(0, 10)
plt.ylim(0, 10)

# Mostrar el gráfico
plt.show()
```
